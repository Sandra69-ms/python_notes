# machine learning


Machine learning is a field of artificial intelligence that enables computers to learn from data without being explicitly programmed.

 It involves using algorithms and statistical models to analyze data, identify patterns, and make predictions or decisions based on those patterns.

In simpler terms, instead of telling a computer exactly how to solve a problem, you give it a lot of data and let it figure out the rules and relationships on its own.

 This allows computers to perform tasks like recognizing images, understanding natural language, making recommendations, and detecting anomalies.


# Types
There are three main types of machine learning:

**Supervised Learning:**

In this type, the algorithm is trained on labeled data, meaning the data includes both the input and the desired output. The goal is to learn a mapping from the input to the output so that the algorithm can predict the output for new, unseen inputs. Examples include classification (e.g., spam detection) and regression (e.g., predicting house prices).

**Unsupervised Learning**:

This type of learning uses unlabeled data. The algorithm tries to find patterns and structures within the data without any prior knowledge of the output. Examples include clustering (grouping similar data points) and dimensionality reduction (reducing the number of features in the data).

**Reinforcement Learning:**

In this type, an agent learns by interacting with an environment. The agent receives rewards or penalties based on its actions, and its goal is to learn a policy that maximizes the cumulative reward over time. Examples include training a robot to walk or developing game-playing AI.

Within supervised learning, classification and regression are two main types of tasks:

**Classification**:

 This is used when the output variable is a category or a label. The goal is to predict which category an input belongs to. For example, classifying emails as "spam" or "not spam," or classifying images of animals into different species.

**Regression**:

 This is used when the output variable is a continuous value. The goal is to predict a numerical value based on the input features. For example, predicting the price of a house based on its size and location, or predicting a person's age based on their physical characteristics.

In essence, classification predicts a discrete outcome, while regression predicts a continuous outcome.

# linear Regression


Linear regression is one of the simplest and most fundamental algorithms in supervised machine learning. It's a type of regression where the goal is to model the relationship between a dependent variable and one or more independent variables by fitting a linear equation to the observed data.

In simple linear regression, where there's only one independent variable, the relationship is represented by a straight line:

y = mx + b

Where:

y is the dependent variable (what you're trying to predict).

x is the independent variable (the input feature).

m is the slope of the line (how much y changes for a one-unit change in x).

b is the y-intercept (the value of y when x is zero).

The algorithm learns the values of m and b that minimize the difference between the predicted y values and the actual y values in the training data.

Linear regression is used for tasks like predicting house prices, sales forecasting, and analyzing trends. It's a good starting point for understanding regression and serves as a basis for more complex models.



 # how is error calculated error measure in linear regression?


In linear regression, the error is essentially the difference between the actual value of the dependent variable and the value predicted by the linear model. This difference is often referred to as the residual.

The goal of linear regression is to find the line that minimizes these residuals. To quantify the overall error of the model, various error measures are used. Some common ones include:

**Mean Absolute Error (MAE):**

This is the average of the absolute differences between the predicted and actual values. It gives an idea of the typical magnitude of the errors.

**Mean Squared Error (MSE):**

This is the average of the squared differences between the predicted and actual values. Squaring the errors gives more weight to larger errors, making it sensitive to outliers.

**Root Mean Squared Error (RMSE):**

 This is the square root of the MSE. It's in the same units as the dependent variable, making it easier to interpret than MSE.

**R-squared (R²):**

This metric represents the proportion of the variance in the dependent variable that is predictable from the independent variables. It indicates how well the model fits the data. A higher R-squared value indicates a better fit.

# data set loading


```python
tip_data = pd.read_excel("/content/tip-amount.xlsx")
tip_data
```

1. Correlation

```python
corrmatrix = tip_data.corr()
corrmatrix
```

2. plot diagram

```python
plt.subplots(figsize= (10,4))
sns.heatmap(corrmatrix, vmin=-.2, vmax =1, annot =True ,linewidths= 0.2 , cmap='YlGnBu')
```

3.Assign data to two variables x and y

```python
y = tip_data["Observed tip amount(yi)"]
x = tip_data.drop(["Observed tip amount(yi)"], axis=1)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, random_state=42, test_size=.25)
```

4. linear regression

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
```

5. fit the model

```python
model = lr.fit(x_train,y_train)
```

6. predict the model


```python
predictions = model.predict(x_test)
```

```python
new_data_point = pd.DataFrame({"Meal": [4,7],"Observed total bill(xi)": [99,150]})
predicted_tip_amount = model.predict(new_data_point)
```

8. Mean squared error

```python
from sklearn.metrics import mean_squared_error
```

```python
print('MSE is : \n ', mean_squared_error(y_test,predictions))
```

9. R-squared test

```python
from sklearn.metrics import r2_score
```

```python
print('R Squared value is : \n ', r2_score(y_test, predictions))
```
